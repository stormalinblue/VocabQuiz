```python
import pandas as pd

words = pd.read_csv('VocabWords.csv')['Word']
words
```

```python
import requests

def get_info_from_api(session, word):
    make_url = lambda word: f'https://api.dictionaryapi.dev/api/v2/entries/en/{word}'
    response = session.get(make_url(word))
    if response.status_code == 404:
        return None
    if response.status_code != 200:
        raise Exception(f'Could not find info on {word}. Response {response.status_code} {response.text}')
    return response.json()
```

```python
responses = {}
```

```python
import time

session = requests.Session()


for word in words:
    if word not in responses:
        print('looking up', word)
        responses[word] = get_info_from_api(session, word)
        time.sleep(5)
        print('got', word)
    else:
        print('skipping', word)
```

```python
import json
with open('words.json', 'w') as f:
    json.dump(responses, f)
```
